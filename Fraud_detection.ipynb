```python
# Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a dataframe
credit_card_df = pd.read_csv('creditcard.csv')
```

```python
credit_card_df.head()
```

```python
credit_card_df.tail()
```

```python
# information about dataset
credit_card_df.info()
```

```python
# distribution of legit transactions and fraudulent transaction
credit_card_df['Class'].value_counts() # this dataset is highly unbalanced
```

0 = Normal transaction
1 = Fraudulent transaction

```python
# seperating the data for analysis
legit = credit_card_df[credit_card_df.Class == 0]
fraud = credit_card_df[credit_card_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_df.groupby('Class').mean()
```

Under-sampling
-> Build a sample dataset containing similar distribution of normal transactoins and the fraudulent transactions  
-> Number of Fraudulent Transaction = 492

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
# concatenate 2 data frames
new_df = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_df.head()
```

```python
new_df.groupby('Class').mean()
```

```python
# split data into features and targets
X = new_df.drop(columns='Class', axis = 1)
Y = new_df['Class']
```

```python
# split the data into training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, 
                                                    stratify=Y, random_state=4)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training -> Using Logistic Regression
model = LogisticRegression()
```

```python
# training the model with training data
model.fit(X_train, y_train)
```

```python
# Predicting
training_predictions = model.predict(X_train)
predictions = model.predict(X_test)
```

```python
# accuracy score
training_accuracy = accuracy_score(training_predictions, y_train)
accuracy = accuracy_score(predictions, y_test)
```

```python
print(training_accuracy) # 93.1% accuracy on training data
```

```python
print(accuracy) # 90.86% accuracy on test data
```
